# Creating Data Frame Using Dictionary

```python
# Creating dataframe using dictionary
data = {'Name':['shivam', 'KArtik', 'Prajjwal', 'ujjwal','daya'],'age':[24,33,22,24,30],'gender':['male','male','male','male','female'],'no':[10,21,31,21,23],'mob':[21313123,1231231,2334223423,1323131,132313]}
df_dict=pd.DataFrame(data)
print(df_dict)
df_dict
```

```python
#to find any one key all value
print(df_dict['Name'])
```

```python
#to select multiple columns in the dataframe
df_dict[['Name','age','mob']]
```

```python
# for update the index
df_dict=pd.DataFrame(data,index=['1-5','2-5','3-5','4-5','5-5'])
```

```python
#filtering our data framae
fil1=df_dict[(df_dict.age>24) & (df_dict.gender=='female')]
print(fil1)
```

```python
fil1=df_dict[df_dict.age>24]
print(fil1)
```

```python
df_dict.age>25
```

```python
# here if we initialize the condition in square brackets than it will be show with data
# and if we initialize the condition without square brackets it will be show true and false only
```

```python
print(df_dict)
```

```python
df_dict.gender=np.where(df_dict.gender=='male','gents','ladies')
print(df_dict)
```

```python
# for filter and replace with any codition expect else are same
df_dict.Name=np.where(df_dict.Name=='KArtik','Kartik Khanna',df_dict.Name)
print(df_dict)
```

```python
#add new column
df_dict['height']=0
print(df_dict)
```

```python
# slicing in data frame
# it can be two types - loc: slicing it by value & iloc: slicing it by index
#iloc first value is row, and second value for column
df_dict.iloc[0:3,0:3]
```

```python
df_dict.loc[df_dict.age>25]
```

```python
# df.head() = display first 5 rows, if we provide the value in df.head() it can give that first number of rows
# df.tail() = display last 5 rows, if we provide the value in df.tail() it can give that last number of rows
# df.info() = display detail about data frame
print('first three rows as per function','\n',df_dict.head(3),'\n')
print('last three rows as per function','\n',df_dict.tail(3),'\n')
df_dict.info()
```

## Renaming the column name

```python
print(df_dict.columns,'orginal name of column','\n')
df_dict.rename(columns={'Name':'Full_Name','age':'Age','gender':'Gender','no':'Lucky_No'}, inplace=True)
print(df_dict.columns,'after replace name of column','\n')
```

```python
#prefix and suffix is used to add start ing with and ending with columns
print(df_dict.columns,'orginal name of column','\n')
df_dict=df_dict.add_prefix('col_')
df_dict=df_dict.add_suffix('_27')
print(df_dict.columns,'after replace name of column','\n')
```
